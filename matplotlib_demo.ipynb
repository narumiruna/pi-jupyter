````python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [v ** 2 for v in x]
print("Generating a matplotlib chart...")
````

````python
fig, ax = plt.subplots(figsize=(5, 3), dpi=120)
ax.plot(x, y, marker="o", linewidth=2, label="y = x²")
ax.bar([v + 0.15 for v in x], [v * 2 for v in x], width=0.25, alpha=0.35, label="2x")
ax.set_title("Matplotlib demo for pi-jupyter")
ax.set_xlabel("x")
ax.set_ylabel("value")
ax.grid(True, alpha=0.3)
ax.legend()
fig.tight_layout()
fig
````
